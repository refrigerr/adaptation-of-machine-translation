```python
import os
import pandas as pd
```

```python
path = "data/witcher_dataset"
```

```python
all_dfs = []

for dirpath, dirnames, filenames in os.walk(path):
    if 'en.csv' in filenames and 'pl.csv' in filenames:
        en_path = os.path.join(dirpath, 'en.csv')
        pl_path = os.path.join(dirpath, 'pl.csv')

        en_df = pd.read_csv(en_path, sep="|", header=1, dtype={'; id      ': str})
        pl_df = pd.read_csv(pl_path, sep="|", header=1, dtype={'; id      ': str})

        en_df.rename(columns={' text': 'en_text'})

        pl_df = pl_df[['; id      ', ' text']].rename(columns={' text': 'pl_text'})

        merged_df = en_df.merge(pl_df, on='; id      ', how='inner')

        all_dfs.append(merged_df)
```

```python
final_df = pd.concat(all_dfs, ignore_index=True)
final_df.head()
```

```python
dataset2 = pd.read_csv("data/witcher_dataset.csv", sep="|")
dataset2
```

```python
dataset2['en_text'] = dataset2['en_text'].str.replace('<i>', '', regex=False)
dataset2['pl_text'] = dataset2['pl_text'].str.replace('<i>', '', regex=False)
```

```python
import json
with open('data/metadata.json', 'r', encoding='utf-8') as file:
    meta_data = json.load(file)
```

```python
meta_data
```

```python
data_by_id = {}
for hex_id, entry in meta_data.items():
    dec_id = int(hex_id, 16)
    data_by_id[dec_id] = {
        'speaker': entry['speaker'],
        'text': entry['text'],
        'next_text': None,
        'prev_text': None
    }

    # Get 'next' text if available
    if 'next' in entry:
        next_id = int(entry['next'][0], 16)
        if meta_data.get(entry['next'][0]):
            data_by_id[dec_id]['next_text'] = meta_data[entry['next'][0]]['text']

    # Get 'prev' text if available
    if 'prev' in entry:
        prev_id = int(entry['prev'][0], 16)
        if meta_data.get(entry['prev'][0]):
            data_by_id[dec_id]['prev_text'] = meta_data[entry['prev'][0]]['text']

df = pd.DataFrame({'id': [int(x, 16) for x in meta_data.keys()]})
```

```python
df
```

```python
df['speaker'] = df['id'].map(lambda x: data_by_id.get(x, {}).get('speaker'))
df['text'] = df['id'].map(lambda x: data_by_id.get(x, {}).get('text'))
df['prev_text'] = df['id'].map(lambda x: data_by_id.get(x, {}).get('prev_text'))
df['next_text'] = df['id'].map(lambda x: data_by_id.get(x, {}).get('next_text'))
```

```python
df
```

```python
merged_df = df.merge(dataset2, on='id', how='inner')
merged_df
```

```python
rows = []

for hex_id, entry in meta_data.items():
    dec_id = int(hex_id, 16)
    row = {
        'id': dec_id,
        'speaker': entry.get('speaker'),
        'text': entry.get('text'),
        'prev_text': None,
        'prev_speaker': None,
        'next_text': None,
        'next_speaker': None
    }

    # Previous
    if 'prev' in entry:
        prev_id_hex = entry['prev'][0]
        prev_entry = meta_data.get(prev_id_hex)
        if prev_entry:
            row['prev_text'] = prev_entry.get('text')
            row['prev_speaker'] = prev_entry.get('speaker')

    # Next
    if 'next' in entry:
        next_id_hex = entry['next'][0]
        next_entry = meta_data.get(next_id_hex)
        if next_entry:
            row['next_text'] = next_entry.get('text')
            row['next_speaker'] = next_entry.get('speaker')

    rows.append(row)

# Create DataFrame
df_json = pd.DataFrame(rows)
df_json
```

```python
merged_df = df_json.merge(dataset2, on='id', how='inner')
```

```python
merged_df
```

```python
merged_df = merged_df.drop(columns=['id', 'text'])
merged_df
```
