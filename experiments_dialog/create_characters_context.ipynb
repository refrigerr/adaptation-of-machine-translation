```python
import pandas as pd
```

```python
df = pd.read_csv("../data/dialog_witcher_dataset.csv", delimiter="|")
df
```

```python
print(df['speaker'].unique())
```

```python
unique_speakers = df['speaker'].unique()
```

```python
def infer_gender(name):
    name_lower = name.lower()

    female_keywords = ['woman', 'girl', 'prostitute', 'lady', 'wife', 'female', 'noblewoman', 'priestess', 'waitress', 'succubus', 'siren', 'sorceress']
    male_keywords = ['man', 'boy', 'dockworker', 'drunk', 'beggar', 'priest', 'guard', 'soldier', 'bard', 'nobleman', 'peasant', 'merchant', 'vendor', 'thug', 'sailor', 'nilfgaardian',
                     'bandit', 'blacksmith', 'villager', 'fisherman', 'boatbuilder', 'warrior', 'berserker', 'captive', 'scribe', 'druid', 'butcher', 'innkeeper', 'hunter',
                     'officer', 'noble', 'medic', 'dwarf', 'cook', 'mage', 'armorer', 'trader', 'scavenger', 'smith', 'man', 'man', 'man']

    if any(keyword in name_lower for keyword in female_keywords):
        return 'female'
    elif any(keyword in name_lower for keyword in male_keywords):
        return 'male'
    else:
        return 'unknown'
```

```python
df['gender'] = df['speaker'].apply(infer_gender)
```

```python
df
```

```python
unique_speakers = df['speaker'].unique()

# Filter speakers where gender is unknown
unknown_gender_speakers = [
    speaker for speaker in unique_speakers
    if infer_gender(speaker) == 'unknown'
]
```

```python
for speaker in unknown_gender_speakers:
    print(speaker)
```

```python
speaker_dict = {}
for speaker in unique_speakers:
    speaker_dict[speaker] = {
        'gender': infer_gender(speaker)
    }
```

```python
import json
```

```python
with open('../data/speakers_characteristics.json', 'w') as f:
    json.dump(speaker_dict, f, indent=4)
```
