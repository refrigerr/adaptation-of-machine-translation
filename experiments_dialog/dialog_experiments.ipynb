```python
import re

def get_match(response):
    match = re.search(r'\[(.*?)\]', response)
    extracted_translation = None
    if match:
        extracted_translation = match.group(1)
    return extracted_translation
```

```python
for index, row in sampled_df.iterrows():
    prompt_dialog_brackets_filled = prompt_dialog_brackets.format(prev_text=row['prev_text'],prev_speaker=row['prev_speaker'],
                                            text=row['en_text'], speaker=row['speaker'],
                                            next_text=row['next_text'], next_speaker=row['next_speaker'])
    
    response = llm.prompt_chat_custom_temperature(prompt=prompt_dialog_brackets_filled, temperature=0.0)
    extracted_response = get_match(response)
    print("eng: ", row['en_text'])
    print("pl: ", row['pl_text'])
    print("transaltion: ", response)
    print("extracted: ", extracted_response)
    print("==================================")
    print()
```

```python
prompt = """ 
Oto kawałek {monolog_or_dialog}{speakers}:

{dialog}

Na podstawie podanego {monolog_or_dialog} oraz kontekstu, przetłumacz tylko i wyłącznie {first_or_second} wypowiedź na język polski, a końcową wersję tłumaczenia zapisz w nawiasach kwadratowych

Oto kilka przykładów jak powinno wyglądać końcowe tłumaczenie w kwadratowych nawiasach:
Zdanie po angielsku: I am a very good cook.
Tłumaczenie: [Jestem bardzo dobrym kucharzem.]

Zdanie po angielsku: Have you heard from her recently?
Tłumaczenie: [Miałeś z nią ostatnio kontakt?]

Zdanie po angielsku: I heard that Michael is going to be a father.
Tłumaczenie: [Słyszałem, że Michael będzie ojcem.]

Na podstawie podanego {monolog_or_dialog} oraz kontekstu, przetłumacz tylko i wyłącznie {first_or_second} wypowiedź na język polski, a końcową wersję tłumaczenia zapisz w nawiasach kwadratowych: {text_to_translate}
"""
```

```python
def choose_prompt(speaker1, speaker2, speaker3):
    if speaker2 == None:
        print("no main speaker")
    elif (speaker1 == None and speaker3 == None):
        monolog_or_dialog = "monologu"
        speakers = f" wypowiadany przez {speaker2}"
        text_to_translate = "TEKST DO PRZETLUMACZENIA"
        dialog = f"{speaker2} - {text_to_translate}"
        first_or_second = "pierwszą"
        print(prompt.format(monolog_or_dialog=monolog_or_dialog, speakers=speakers, dialog=dialog, first_or_second=first_or_second, text_to_translate=text_to_translate))
    elif (speaker2 == speaker1 and speaker3 == None):
        monolog_or_dialog = "monologu"
        speakers = f" wypowiadany przez {speaker2}"
        previous_text = "jakis poprzedni tekst"
        text_to_translate = "TEKST DO PRZETLUMACZENIA"
        dialog = f"{speaker1} - {previous_text} \n{speaker2} - {text_to_translate}"
        first_or_second = "drugą"
        print(prompt.format(monolog_or_dialog=monolog_or_dialog, speakers=speakers, dialog=dialog, first_or_second=first_or_second, text_to_translate=text_to_translate))
    elif (speaker2 == speaker3 and speaker1 == None):
        monolog_or_dialog = "monologu"
        speakers = f" wypowiadany przez {speaker2}"
        text_to_translate = "TEKST DO PRZETLUMACZENIA"
        next_text = "jakis nastepny tekst"
        dialog = f"{speaker2} - {text_to_translate} \n{speaker3} - {next_text}"
        first_or_second = "pierwszą"
        print(prompt.format(monolog_or_dialog=monolog_or_dialog, speakers=speakers, dialog=dialog, first_or_second=first_or_second, text_to_translate=text_to_translate))
    elif (speaker2 == speaker1 and speaker2 == speaker3):
        monolog_or_dialog = "monologu"
        speakers = f" wypowiadany przez {speaker2}"
        previous_text = "jakis poprzedni tekst"
        text_to_translate = "TEKST DO PRZETLUMACZENIA"
        next_text = "jakis nastepny tekst"
        dialog = f"{speaker1} - {previous_text} \n{speaker2} - {text_to_translate} \n{speaker3} - {next_text}"
        first_or_second = "drugą"
        print(prompt.format(monolog_or_dialog=monolog_or_dialog, speakers=speakers, dialog=dialog, first_or_second=first_or_second, text_to_translate=text_to_translate))
    elif (speaker1 != speaker2 and speaker1 != None and speaker3 == None):
        monolog_or_dialog = "dialogu"
        speakers = f", rozmawia w nim {speaker2} z {speaker1}"
        previous_text = "jakis poprzedni tekst"
        text_to_translate = "TEKST DO PRZETLUMACZENIA"
        dialog = f"{speaker1} - {previous_text} \n{speaker2} - {text_to_translate}"
        first_or_second = "drugą"
        print(prompt.format(monolog_or_dialog=monolog_or_dialog, speakers=speakers, dialog=dialog, first_or_second=first_or_second, text_to_translate=text_to_translate))
    elif (speaker3 != speaker2 and speaker3 != None and speaker1 == None):
        monolog_or_dialog = "dialogu"
        speakers = f", rozmawia w nim {speaker2} z {speaker3}"
        text_to_translate = "TEKST DO PRZETLUMACZENIA"
        next_text = "jakis nastepny tekst"
        dialog = f"{speaker2} - {text_to_translate} \n{speaker3} - {next_text}"
        first_or_second = "pierwszą"
        print(prompt.format(monolog_or_dialog=monolog_or_dialog, speakers=speakers, dialog=dialog, first_or_second=first_or_second, text_to_translate=text_to_translate))
    elif (speaker1 != speaker2 and speaker1 != None and speaker3 != None) or (speaker3 != speaker2 and speaker3 != None and speaker1 != None):
        monolog_or_dialog = "dialogu"

        if speaker1 == speaker2 and speaker1 != speaker3:
            speakers = f", rozmawia w nim {speaker2} z {speaker3}"
        elif speaker3 == speaker2 and speaker3 != speaker1:
            speakers = f", rozmawia w nim {speaker2} z {speaker1}"
        elif speaker1 == speaker3 and speaker1 != speaker2:
            speakers = f", rozmawia w nim {speaker2} z {speaker1}"
        else:
            speakers = f", rozmawia w nim {speaker2} z {speaker1} oraz z {speaker3}"

        previous_text = "jakis poprzedni tekst"
        text_to_translate = "TEKST DO PRZETLUMACZENIA"
        next_text = "jakis nastepny tekst"
        dialog = f"{speaker1} - {previous_text} \n{speaker2} - {text_to_translate} \n{speaker3} - {next_text}"
        first_or_second = "drugą"
        print(prompt.format(monolog_or_dialog=monolog_or_dialog, speakers=speakers, dialog=dialog, first_or_second=first_or_second, text_to_translate=text_to_translate))
    else:
        print("error")
```
