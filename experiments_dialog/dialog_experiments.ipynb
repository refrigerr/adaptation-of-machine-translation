```python
import pandas as pd
import sys
sys.path.append('..')
```

```python
from llm_class import LLM

llm = LLM(model_id="bielik", key_path='../key.txt')
```

```python
df = pd.read_csv("../data/dialog_witcher_dataset.csv", delimiter="|")
```

```python
df
```

```python
def not_blank(s):
    return s.notnull() & s.str.strip().ne('')
```

```python
filtered_df = df[not_blank(df['prev_text']) & not_blank(df['next_text'])]
filtered_df
```

```python
sampled_df = filtered_df.sample(n=5, random_state=10).reset_index(drop=True)
sampled_df
```

```python
prompt_default = """
Przetłumacz na polski i nie dodawaj żadnych dodatkowych wyjaśnień, napisz tylko samo tłumaczenie: {en_text}
"""
```

```python
prompt_dialog = """ 
Oto kawałek dialogu:

- {prev_text}
- {text}
- {next_text}

Na podstawie podanego dialogu, przetłumacz tylko i wyłącznie drugą wypowiedź na język polski i nie dodawaj żadnych dodatkowych wyjaśnień, napisz tylko samo tłumaczenie: {text}
"""
```

```python
prompt_dialog2 = """ 
Oto kawałek dialogu:

{prev_speaker}: {prev_text}
{speaker}: {text}
{next_speaker}: {next_text}

Na podstawie podanego dialogu, przetłumacz tylko i wyłącznie drugą wypowiedź na język polski i nie dodawaj żadnych dodatkowych wyjaśnień, napisz tylko samo tłumaczenie: '{text}'
"""
```

```python
for index, row in sampled_df.iterrows():
    prompt_filled = prompt_dialog.format(prev_text=row['prev_text'], text=row['en_text'], next_text=row['next_text'])
    response = llm.prompt_chat_custom_temperature(prompt=prompt_filled, temperature=0.0)

    print("eng: ", row['en_text'])
    print("pl: ", row['pl_text'])
    print("transaltion: ", response)
    print("==================================")
    print()
```

```python
for index, row in sampled_df.iterrows():
    prompt_filled = prompt_dialog2.format(prev_text=row['prev_text'],prev_speaker=row['prev_speaker'],
                                           text=row['en_text'], speaker=row['speaker'],
                                           next_text=row['next_text'], next_speaker=row['next_speaker'])
    response = llm.prompt_chat_custom_temperature(prompt=prompt_filled, temperature=0.0)

    print("eng: ", row['en_text'])
    print("pl: ", row['pl_text'])
    print("transaltion: ", response)
    print("==================================")
    print()
```

```python
sampled_df = filtered_df.sample(n=1000, random_state=10).reset_index(drop=True)
```

```python
for index, row in sampled_df.iterrows():

    prompt_default_filled = prompt_default.format(en_text=row['en_text'])
    prompt_dialog_filled = prompt_dialog.format(prev_text=row['prev_text'], text=row['en_text'], next_text=row['next_text'])
    prompt_dialog2_filled = prompt_dialog2.format(prev_text=row['prev_text'],prev_speaker=row['prev_speaker'],
                                           text=row['en_text'], speaker=row['speaker'],
                                           next_text=row['next_text'], next_speaker=row['next_speaker'])
    
    
    response_default = llm.prompt_chat_custom_temperature(prompt=prompt_default_filled, temperature=0.0)
    response_dialog = llm.prompt_chat_custom_temperature(prompt=prompt_dialog_filled, temperature=0.0)
    response_dialog2 = llm.prompt_chat_custom_temperature(prompt=prompt_dialog2_filled, temperature=0.0)

    sampled_df.at[index, f'default'] = response_default
    sampled_df.at[index, f'dialog'] = response_dialog
    sampled_df.at[index, f'dialog2'] = response_dialog2

sampled_df.to_csv("results/dialog_experiments.csv", sep="|")
```

```python
prompt_dialog_brackets = """ 
Oto kawałek dialogu:

- {prev_text}
- {text}
- {next_text}

Na podstawie podanego dialogu, przetłumacz tylko i wyłącznie drugą wypowiedź na język polski, a końcową wersję tłumaczenia zapisz w nawiasach kwadratowych

Oto kilka przykładów jak powinno wyglądać końcowe tłumaczenie w kwadratowych nawiasach:
Zdanie po angielsku: I am a very good cook.
Tłumaczenie: [Jestem bardzo dobrym kucharzem.]

Zdanie po angielsku: Have you heard from her recently?
Tłumaczenie: [Miałeś z nią ostatnio kontakt?]

Zdanie po angielsku: I heard that Michael is going to be a father.
Tłumaczenie: [Słyszałem, że Michael będzie ojcem.]

Na podstawie podanego dialogu, przetłumacz tylko i wyłącznie drugą wypowiedź na język polski, a końcową wersję tłumaczenia zapisz w nawiasach kwadratowych: {text}
"""
```

```python
sampled_df = filtered_df.sample(n=5, random_state=10).reset_index(drop=True)
```

```python
for index, row in sampled_df.iterrows():
    prompt_dialog_brackets_filled = prompt_dialog_brackets.format(prev_text=row['prev_text'],prev_speaker=row['prev_speaker'],
                                            text=row['en_text'], speaker=row['speaker'],
                                            next_text=row['next_text'], next_speaker=row['next_speaker'])
    
    response = llm.prompt_chat_custom_temperature(prompt=prompt_dialog_brackets_filled, temperature=0.0)

    print("eng: ", row['en_text'])
    print("pl: ", row['pl_text'])
    print("transaltion: ", response)
    print("==================================")
    print()
```

```python
import re

def get_match(response):
    match = re.search(r'\[(.*?)\]', response)
    extracted_translation = None
    if match:
        extracted_translation = match.group(1)
    return extracted_translation
```

```python
for index, row in sampled_df.iterrows():
    prompt_dialog_brackets_filled = prompt_dialog_brackets.format(prev_text=row['prev_text'],prev_speaker=row['prev_speaker'],
                                            text=row['en_text'], speaker=row['speaker'],
                                            next_text=row['next_text'], next_speaker=row['next_speaker'])
    
    response = llm.prompt_chat_custom_temperature(prompt=prompt_dialog_brackets_filled, temperature=0.0)
    extracted_response = get_match(response)
    print("eng: ", row['en_text'])
    print("pl: ", row['pl_text'])
    print("transaltion: ", response)
    print("extracted: ", extracted_response)
    print("==================================")
    print()
```

```python
sampled_df = filtered_df.sample(n=1000, random_state=10).reset_index(drop=True)
```

```python
for index, row in sampled_df.iterrows():
    prompt_dialog_brackets_filled = prompt_dialog_brackets.format(prev_text=row['prev_text'],prev_speaker=row['prev_speaker'],
                                            text=row['en_text'], speaker=row['speaker'],
                                            next_text=row['next_text'], next_speaker=row['next_speaker'])
    
    response = llm.prompt_chat_custom_temperature(prompt=prompt_dialog_brackets_filled, temperature=0.0)

    extracted_translation = get_match(response=response)

    sampled_df.at[index, f'dialog_extracted'] = extracted_translation
    
    if (index + 1) % 100 == 0:
        print(f"Processed {index + 1} out of {len(sampled_df)} samples...")

sampled_df.to_csv("results/dialog_extraction_experiments.csv", sep="|", index=False)
```
