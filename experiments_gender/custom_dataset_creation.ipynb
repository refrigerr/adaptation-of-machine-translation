```python
import pandas as pd

data = []
entry = {}

with open('../examples/gender_examples_dataset.txt', 'r', encoding='utf-8') as file:
    for line in file:
        line = line.strip()
        if not line:
            continue
        
        if ':' in line:
            key, value = line.split(':', 1)
            key = key.strip()
            value = value.strip()
            
            if key == 'Mówca':
                entry['speaker_gender'] = value
            elif key == 'Słuchacz':
                entry['listener_gender'] = value
            elif key == 'Angielski':
                entry['en_text'] = value
            elif key == 'Polski':
                entry['pl_text'] = value

        if 'pl_text' in entry:
            data.append(entry)
            entry = {}

df = pd.DataFrame(data)

df
```

```python
df.to_csv('../data/custom_gender_examples_dataset.csv', index=False, sep='|')
```
