```python
import pandas as pd
from llm_class import LLM
```

```python
llm = LLM(model_id='bielik', key_path="../key.txt")
```

```python
dataset = pd.read_csv('../data/custom_gender_examples_dataset.csv', sep='|')
dataset
```

```python
few_shot_examples = pd.read_csv('../data/custom_gender_examples_few_shot.csv', sep='|')
few_shot_examples
```

```python
prompt = """
Przetłumacz na polski i nie dodawaj żadnych dodatkowych wyjaśnień, napisz tylko samo tłumaczenie: {en_text}
"""
```

```python
prompt2 = """
Zdanie jest częscią dialogu. Wypowiada je {speaker_gender} w stronę {listener_gender}.

Przetłumacz na polski według podanego kontekstu i nie dodawaj żadnych dodatkowych wyjaśnień, napisz tylko samo tłumaczenie: {en_text}
"""
```

```python
prompt3 = """
Oto kontekst zdania: Zdanie jest częscią dialogu. Wypowiada je {speaker_gender} w stronę {listener_gender}.

Przetłumacz na polski według podanego kontekstu i nie dodawaj żadnych dodatkowych wyjaśnień, napisz tylko samo tłumaczenie: {en_text}
"""
```

```python
prompt4 = """
Mówca: {speaker_gender}
Słuchacz: {listener_gender}
Angielski: {en_text}
Polski:
"""
```

```python
prompt5 = """
Mówca: {speaker_gender}
Słuchacz: {listener_gender}

Przetłumacz na polski według podanego kontekstu i nie dodawaj żadnych dodatkowych wyjaśnień, napisz tylko samo tłumaczenie: {en_text}
"""
```

```python
def sample_random_examples(n=5, random_state=42):
    sampled_rows = few_shot_examples.sample(5, random_state=random_state)

    formatted_text = ""

    for _, row in sampled_rows.iterrows():
        formatted_text += f"""
Mówca: {row['speaker_gender']}
Słuchacz: {row['listener_gender']}
Angielski: {row['en_text']}
Polski: {row['pl_text']}
"""
    return formatted_text
```

```python
print(sample_random_examples())
```

```python
def sample_examples_by_gender(df, speaker_gender, listener_gender, n=5):
   
    filtered_df = df[(df['speaker_gender'] == speaker_gender) & (df['listener_gender'] == listener_gender)]

    sampled_rows = filtered_df.sample(n, random_state=1)

    formatted_text = ""

    for _, row in sampled_rows.iterrows():
        formatted_text += f"""
Mówca: {row['speaker_gender']}
Słuchacz: {row['listener_gender']}
Angielski: {row['en_text']}
Polski: {row['pl_text']}
"""

    return formatted_text
```

```python
print(sample_examples_by_gender(few_shot_examples, "mężczyzna", "kobieta"))
```

```python
dataset['prompt1'] = None
for index, row in dataset.iterrows():

    prompt_filled = prompt.format(en_text=row['en_text'])

    result = llm.prompt_chat_custom_temperature(prompt=prompt_filled, temperature=0.0)
    dataset.at[index, 'prompt1'] = result

dataset['prompt2'] = None

for index, row in dataset.iterrows():
    speaker_gender = row['speaker_gender']
    listener_gender = "mężczyzny" if row['listener_gender'] == "mężczyzna" else "kobiety"
    en_text = row['en_text']

    prompt_filled = prompt2.format(en_text=row['en_text'], speaker_gender=speaker_gender, listener_gender=listener_gender)

    result = llm.prompt_chat_custom_temperature(prompt=prompt_filled, temperature=0.0)
    dataset.at[index, 'prompt2'] = result

dataset['prompt3'] = None

for index, row in dataset.iterrows():
    speaker_gender = row['speaker_gender']
    listener_gender = "mężczyzny" if row['listener_gender'] == "mężczyzna" else "kobiety"
    en_text = row['en_text']

    prompt_filled = prompt3.format(en_text=row['en_text'], speaker_gender=speaker_gender, listener_gender=listener_gender)

    result = llm.prompt_chat_custom_temperature(prompt=prompt_filled, temperature=0.0)
    dataset.at[index, 'prompt3'] = result

dataset['prompt4'] = None

for index, row in dataset.iterrows():
    speaker_gender = row['speaker_gender']
    listener_gender = row['listener_gender']
    en_text = row['en_text']

    prompt_filled = prompt4.format(en_text=row['en_text'], speaker_gender=speaker_gender, listener_gender=listener_gender)

    result = llm.prompt_chat_custom_temperature(prompt=prompt_filled, temperature=0.0)
    dataset.at[index, 'prompt4'] = result

dataset['prompt5'] = None

for index, row in dataset.iterrows():
    speaker_gender = row['speaker_gender']
    listener_gender = row['listener_gender']
    en_text = row['en_text']

    prompt_filled = prompt5.format(en_text=row['en_text'], speaker_gender=speaker_gender, listener_gender=listener_gender)

    result = llm.prompt_chat_custom_temperature(prompt=prompt_filled, temperature=0.0)
    dataset.at[index, 'prompt5'] = result
```

```python
dataset.to_csv('results/custom_dataset_prompts_comparision.csv', sep='|', index=False)
```

```python
prompt6 = """
Oto kilka przykładowych tłumaczeń:
{examples}

Mówca: {speaker_gender}
Słuchacz: {listener_gender}

Przetłumacz na polski według podanego kontekstu oraz przykładów i nie dodawaj żadnych dodatkowych wyjaśnień, napisz tylko samo tłumaczenie: {en_text}
"""
```

```python
prompt7 = """
Oto kontekst zdania: Zdanie jest częscią dialogu. Wypowiada je {speaker_gender} w stronę {listener_gender}.

Oto kilka przykładowych tłumaczeń:
{examples}

Przetłumacz na polski według podanego kontekstu oraz przykładów i nie dodawaj żadnych dodatkowych wyjaśnień, napisz tylko samo tłumaczenie: {en_text}
"""
```

```python
dataset['prompt6_random'] = None
for index, row in dataset.iterrows():

    speaker_gender = row['speaker_gender']
    listener_gender = row['listener_gender']
    en_text = row['en_text']

    examples = sample_random_examples(n=5, random_state=index)

    prompt_filled = prompt6.format(en_text=row['en_text'], speaker_gender=speaker_gender,
                                    listener_gender=listener_gender, examples=examples)

    result = llm.prompt_chat_custom_temperature(prompt=prompt_filled, temperature=0.0)
    dataset.at[index, 'prompt6_random'] = result

dataset['prompt6_chosen'] = None

for index, row in dataset.iterrows():
    speaker_gender = row['speaker_gender']
    listener_gender = row['listener_gender']
    en_text = row['en_text']

    examples = sample_examples_by_gender(df=few_shot_examples, speaker_gender=speaker_gender,
                                          listener_gender=listener_gender, n=5)

    prompt_filled = prompt6.format(en_text=row['en_text'], speaker_gender=speaker_gender,
                                    listener_gender=listener_gender, examples=examples)

    result = llm.prompt_chat_custom_temperature(prompt=prompt_filled, temperature=0.0)
    dataset.at[index, 'prompt6_chosen'] = result

dataset['prompt7_random'] = None

for index, row in dataset.iterrows():
    speaker_gender = row['speaker_gender']
    listener_gender = row['listener_gender']
    en_text = row['en_text']

    examples = sample_random_examples(n=5, random_state=index)

    listener_gender = "mężczyzny" if row['listener_gender'] == "mężczyzna" else "kobiety"

    prompt_filled = prompt7.format(en_text=row['en_text'], speaker_gender=speaker_gender,
                                    listener_gender=listener_gender, examples=examples)

    result = llm.prompt_chat_custom_temperature(prompt=prompt_filled, temperature=0.0)
    dataset.at[index, 'prompt7_random'] = result

dataset['prompt7_chosen'] = None

for index, row in dataset.iterrows():
    speaker_gender = row['speaker_gender']
    listener_gender = row['listener_gender']
    en_text = row['en_text']

    examples = sample_examples_by_gender(df=few_shot_examples, speaker_gender=speaker_gender,
                                          listener_gender=listener_gender, n=5)
    
    listener_gender = "mężczyzny" if row['listener_gender'] == "mężczyzna" else "kobiety"

    prompt_filled = prompt7.format(en_text=row['en_text'], speaker_gender=speaker_gender, 
                                   listener_gender=listener_gender, examples=examples)

    result = llm.prompt_chat_custom_temperature(prompt=prompt_filled, temperature=0.0)
    dataset.at[index, 'prompt7_chosen'] = result
```

```python
dataset.to_csv('results/custom_dataset_few_shot_comparision.csv', sep='|', index=False)
```
